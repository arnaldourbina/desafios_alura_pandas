# **Leyendo BD proyecto desafio 1**

```python
import pandas as pd

ventas_clientes = pd.read_json('/content/dados_vendas_clientes.json')

ventas_clientes.head()
```

```python
columnas = list(ventas_clientes.columns)
columnas
```

### **Normalizando**

```python
ventas_clientes_norm = pd.json_normalize(ventas_clientes['dados_vendas'])
ventas_clientes_norm
```

### **Explotando listas**

```python
columnas = list(ventas_clientes_norm.columns)
columnas
```

```python
ventas_clientes_exp = ventas_clientes_norm.explode(columnas[1:])
```

```python
ventas_clientes_exp.reset_index(inplace=True, drop=True)
```

```python
ventas_clientes_exp
```

### **Verificando tipos de datos**

```python
ventas_clientes_exp.info()
```

### **Identificando columnas numericas**

Valor da compra

```python
columna_numerica = ventas_clientes_exp['Valor da compra']
```

### **Transformar columnas numericas en numericas**

```python
import numpy as np
```

```python
from os import remove
ventas_clientes_exp['Valor da compra'] = ventas_clientes_exp['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip())
```

```python
ventas_clientes_exp
```

```python
ventas_clientes_exp['Valor da compra'] = ventas_clientes_exp['Valor da compra'].astype(float)
```

```python
ventas_clientes_exp.info()
```

### **Limpiando los datos de columna cliente**

```python
ventas_clientes_exp['Cliente'] = ventas_clientes_exp['Cliente'].str.lower()
```

```python
ventas_clientes_exp
```

```python
ventas_clientes_exp['Cliente'] = ventas_clientes_exp['Cliente'].str.replace('ã','a')
```

```python
ventas_clientes_exp
```

### manipula los textos de la columna Cliente para que el resultado sean los nombres de los clientes en letras minúsculas, con ausencia de caracteres especiales o números.

```python
ventas_clientes_exp['Cliente'] = ventas_clientes_exp['Cliente'].str.replace('[^a-zA-Z]',' ',regex=True)
```
